```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("./img1.JPG")
plt.imshow(img1[:,:,::-1])
plt.show()

img2 = cv2.imread("./img2.jpg")
plt.imshow(img2[:,:,::-1])
plt.show()
```

```python
result = DeepFace.verify(img1, img2)
print("The results are: ", result)
```
